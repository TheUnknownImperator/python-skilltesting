Things needed:

*   Importing relevant libraries
*   Entering the URL

After that, use pd.read_html(. . .) to read tables from the received response.

```python
import requests
import pandas as pd

URL = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area"
response = pd.read_html(URL)
```

Pick the table using the appropriate index. As the table hasn't looked exactly as one wanted, edit the table a bit more using:

*   .drop() function to remove certain rows or columns
*   .reset_index function to restart indexing from 0
*   .iloc[a:b,x:y] to locate and pick specific parts based on index

among others.

```python
table = response[1]
table = table.drop([0,2], axis = 0).drop(['Unnamed: 0','Unnamed: 6'], axis=1)
table = table.reset_index()
table = table.drop('index',axis=1)
table.index = table.index + 1
top50 = table.iloc[0:50,:]
```

As one have located the top 50 on the table, one can finally print the finished list.

```python
print("Top 50 nations in terms of largest territories on Earth: ")
top50
```
