Things needed:
*   Importing relevant libraries
*   Entering the URL
*   Defining path for usage

```python
import os
import requests
from PIL import Image
URL = "https://en.touhouwiki.net/images/d/d7/Th15front.jpg"

path = os.getcwd()
paths = os.path.join(path,'kanjudentesting.jpg')
```

Next step: requesting from the URL and making sure the filetype requested is correct

```python
r = requests.get(URL)
print(r.headers['Content-Type'])
```

After that: enter the requested content into the declared path

```python
with open(paths, 'wb') as ftype:
  ftype.write(r.content)
```

Lastly, use Image.open(. . .) to show the image

```python
Image.open(paths)
```
